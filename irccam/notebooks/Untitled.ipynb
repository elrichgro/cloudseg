```python
import sys, os
sys.path.append(os.path.join(os.path.abspath(''), '..'))
sys.path.append(os.path.join(os.path.abspath(''), '../..'))
import numpy as np
import cv2
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
%matplotlib inline
from datasets.cloud_dataset import HDF5Dataset
import h5py
```

```python
h5file=h5py.File("../../data/datasets/test3/20180105.h5");

def get(index):
    return h5file['irc'][index],h5file['vis'][index],h5file['clearsky'][index]
def plot_irc(img):
    figure(num=None, figsize=(16, 12), dpi=80, facecolor='w', edgecolor='k')
    implot = plt.imshow(img, cmap='gray')
def plot_rgb(img):
    figure(num=None, figsize=(16, 12), dpi=80, facecolor='w', edgecolor='k')
    implot = plt.imshow(img / 255)
```

```python
irc, rgb, clear = get(12)

plot_irc(irc)
plot_rgb(rgb)

```
